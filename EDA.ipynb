```python
dfType1 = df[df['A&E attendances Type 1'] != 0]
```

```python
dfType1['Type 1 Proportion attendances over 4hrs'] = dfType1['Attendances over 4hrs Type 1']/dfType1['A&E attendances Type 1']
```

```python
dfType1['Total Admissions'] = dfType1['Emergency admissions via A&E - Type 1'] + dfType1['Emergency admissions via A&E - Type 2'] + dfType1['Emergency admissions via A&E - Other A&E department']
```

```python
dfType1['Proportion waiting over 4hrs for admission'] = (dfType1['Patients who have waited 4-12 hs from DTA to admission'] + dfType1['Patients who have waited 12+ hrs from DTA to admission'])/dfType1['Total Admissions']
dfType1['Proportion waiting over 12hrs for admission'] = (dfType1['Patients who have waited 12+ hrs from DTA to admission'])/dfType1['Total Admissions']
```

```python
# Drop TOTAL row
dfType1 = dfType1[:-1]
```

```python
dfType1[['Parent Org','Org name','A&E attendances Type 1','Total Admissions','Type 1 Proportion attendances over 4hrs','Proportion waiting over 4hrs for admission','Proportion waiting over 12hrs for admission']].to_csv('performance_metrics.csv')
```
